# Basic Visualization

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_canada = pd.read_excel("./data/Canada.xlsx", sheet_name="Canada by Citizenship",
                          skiprows=range(20), skipfooter=2)
print("Num of rows:", df_canada.shape[0])
print("Num of cols:", df_canada.shape[1])
```

```python
display(df_canada.head())
display(df_canada.tail())
```

```python
df_canada.columns
```

```python
df_canada = df_canada.drop(columns=["Type", "Coverage", "AREA", "REG", "DEV"])
df_canada.head()
```

```python
df_canada = df_canada.rename(columns={
    "OdName": "Country",
    "RegName": "Region",
    "AreaName": "Continent",
})
df_canada.head()
```

```python
df_canada["Total"] = df_canada.sum(axis=1)
df_canada.head()
```

```python
df_canada.describe()
```

```python
df_canada[["Country", 2012, 2013]]
```

```python
(df_canada.Country.value_counts() > 1).sum()
```

```python
df_canada = df_canada.set_index("Country")
```

```python
df_canada.index.name = None
```

## Plotting using Matplotlib

```python
print("matplotlib version:", matplotlib.__version__)
```

```python
plt.style.use("fivethirtyeight")
```

```python
indonesia = df_canada.loc["Indonesia", years]
indonesia
```

```python
indonesia.plot()
plt.title("Immigrants from Indonesia")
plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
plt.text(1993, 1000, "1998 Crisis", fontsize=10)
plt.show()
```

```python
imci = df_canada.loc[["Indonesia", "Malaysia", "China", "India"], years]
imci
```

```python
imci.T.plot(figsize=(12, 6))
plt.title("Immigrants from Indonesia, Malaysia, China, India")
plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
plt.show()
```

> Top **5 countries** that contributed the most to immigrants to Canada

```python
top_5_countries = df_canada.sort_values(by="Total", ascending=False).head(n=5)[years]
top_5_countries
```

```python
top_5_countries.T.plot(figsize=(12, 7))
plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
plt.title("Top 5 countries")
plt.show()
```

```python
top_5_countries.T.plot(
    kind="area", figsize=(13, 7),
)
plt.show()
```

```python
top_5_countries.T.plot(
    kind="area", figsize=(13, 7),
    stacked=False, alpha=.35
)
plt.savefig("./data/top5_countries.png")
plt.show()
```

```python
df_canada[2013].plot(kind="hist", figsize=(12, 6), bins=10)
plt.title("Immigrants Distribution from various countries in 2013")
plt.ylabel("Number of countries")
plt.xlabel("Number of immigrants")
plt.show()
```

```python
dns = df_canada.loc[["Denmark", "Norway", "Sweden"], years]
dns
```

```python
_, bin_edges = np.histogram(dns, bins=15)
dns.T.plot(
    kind="hist",
    figsize=(14, 7),
    alpha=.6,
    bins=15,
    xticks=bin_edges,
    color=["coral", "darkslateblue", "mediumseagreen"]
)
plt.title("Immigrants Distribution from Denmark, Norway, and Sweden")
plt.xlabel("Number of Immigrants")
plt.ylabel("Number of Countries")
plt.show()
```

```python
iceland = df_canada.loc["Iceland", years]
iceland
```

```python
iceland.plot(kind="bar", figsize=(10, 6))
plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
plt.title("Iceland Immigrants from 1980 - 2013")

plt.annotate("", xy=(32.5, 70), xytext=(28, 20), xycoords="data",
             arrowprops={"arrowstyle": "->",
                         "connectionstyle": "arc3",
                         "color": "darkblue", "lw": 2})

plt.annotate("2008-2011 Financial Crisis", xy=(28, 30), rotation=70,
             va="bottom", ha="left")

plt.show()
```

```python
top_10_countries_total = df_canada.sort_values("Total", ascending=False).head(10)["Total"]
top_10_countries_total
```

```python
top_10_countries_total.plot(kind="barh", figsize=(10, 11))
plt.xlabel("Number of Immigrants")
plt.ylabel("Country")
plt.title("Top 5 Countries Contributed to The Most Immigrants in Canada 1980 - 2013")
plt.show()
```

```python
continents = df_canada.groupby("Continent").sum()

pie_colors = ["gold", "yellowgreen", "lightcoral", "lightskyblue", "lightgreen", "pink"]
explode_list = [.1, 0, 0, 0, .1, .1]
continents["Total"].plot(kind="pie", figsize=(13, 6), shadow=True,
                         autopct="%1.1f%%", explode=explode_list,
                         colors=pie_colors, pctdistance=1.15, labels=None)
plt.legend(labels=continents.index, loc="best")
```

```python
indonesia.plot(kind="box")
plt.title("Indonesian Immigrants from 1980 to 2013")
plt.ylabel("Number of Immigrants")
plt.show()
```

```python
imci.loc[["China", "India"]].T.plot(kind="box", figsize=(9, 5),
                                    vert=False, patch_artist=True)
```

```python
imci.loc[["Indonesia", "Malaysia"]].T.plot(kind="box", figsize=(9, 5))
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.set_title("Immigrants from China and India")
ax1.set_ylabel("Nummber of Immigrants")
ax1.set_xlabel("Country")
ax2.set_title("Immigrants from Indonesia and Malaysia")
ax2.set_xlabel("Country")
plt.suptitle("Immigrants Contribution", y=1.)

imci.loc[["Indonesia", "Malaysia"]].T.plot(kind="area", alpha=.4, ax=ax2)
imci.loc[["China", "India"]].T.plot(kind="area", alpha=.65, ax=ax1)
plt.savefig("./data/immigrants.png")
plt.show()
```

```python
id_ma = imci.loc[["Indonesia", "Malaysia"]]
cn_in = imci.loc[["China", "India"]]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
for df, ax in zip((id_ma, cn_in), axes):
    ax.set_title(
        f"{' and '.join(df.index.tolist())} Immigrants"
    )
    df.T.plot(kind="line", ax=ax)
```
